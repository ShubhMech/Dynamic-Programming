```python
def memoization(arr,table,val,wt,n):
    if n<= 0 or wt <= 0:
        return 0
    
    elif wt-arr[n]<0:
        if table[wt][n] == -1:
            table[wt][n] = memoization(arr,table,val,wt,n-1)
            return table[wt][n]
        else:
            return table[wt][n]
    
    else:
        table[wt][n] = max(val[n] + memoization(arr,table,val,wt-arr[n],n-1), memoization(arr,table,val,wt,n-1))
        return table[wt][n]
    
    
```

```python
arr= [12,13,14,5,3,16,14]
n= len(arr)
wt = 56
val=[111,132,132,145,123,145,119]
table = [[-1 for x in range(0,n+1)] for j in range(0,wt+1)]
```

```python
memoization(arr,table,val,wt,n-1)
```
